```python
import numpy as np
from numpy import log10
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import lfilter
from sklearn.metrics import mean_squared_error
```

```python
# Load the EEG signal dataset from a CSV file
eeg_data = pd.read_csv('eeg_data.csv')
```

```python
signal = eeg_data['EEG.AF3'].values
```

```python
# Plot the noisy signal
plt.figure()
plt.plot(signal)
plt.title('Original Signal')
```

```python
# Add Poisson noise to the signal
lambda_ = 5  # parameter of the Poisson distribution
noisy_signal = np.random.poisson(lambda_ * np.abs(signal)) * np.sign(signal)
```

```python
# Plot the noisy signal
plt.figure()
plt.plot(noisy_signal)
plt.title('Noisy signal')
```

```python
# Define the LMS adaptive filter
filter_order = 3
mu = 0.000000001
epsilon = 1e-5 # small constant to avoid division by zero
weights = 0.01 * np.random.randn(filter_order)
```

```python
# Filter the noisy signal using the NLMS adaptive filter
filtered_signal = np.zeros(noisy_signal.shape)
for i in range(filter_order, len(noisy_signal)):
    x = noisy_signal[i-filter_order:i]
    y = np.dot(weights, x)
    error = signal[i] - y
    norm_x = np.linalg.norm(x) + epsilon
    weights += (mu * error * x) / norm_x
    filtered_signal[i] = y
```

```python
mse = mean_squared_error(noisy_signal, filtered_signal)
psnr = 20 * log10(max(noisy_signal)) - 10 * log10(mse)
snr = 10 * log10(np.sum(noisy_signal**2) / np.sum((noisy_signal - filtered_signal)**2))
```

```python
# Plot the original, noisy, and filtered signals
plt.plot(signal, label='Original signal')
plt.plot(noisy_signal, label='Noisy signal')
plt.plot(filtered_signal, label='Filtered signal')
plt.legend()
plt.show()

# Print the PSNR, MSE, and SNR values
print("PSNR:", psnr)
print("MSE:", mse)
print("SNR:", snr)
```
